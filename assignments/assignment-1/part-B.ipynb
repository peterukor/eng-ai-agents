This assignment asks if it's better to make one model for all houses, or a separate model for each neighborhood. I'll walk through what I found.

First I tried one single model for all the houses, rich and poor together.

The problem is rich houses go up in price a lot faster per square foot than poor houses do. Rich houses go up about $350 per extra square foot. Poor houses only go up about $120 per square foot. One model only gets to pick one slope, so it has to kind of average between the two, which makes it wrong for both groups.

Here is what I got:

one global model, test error (RMSE) = $91,347
two separate models, one for rich one for poor, test error (RMSE) = $23,069

So splitting into two models made the error almost 4 times smaller.

Next question, what if I just add a column called is_rich (0 or 1) to the one big model instead of splitting it into two models?

My guess before running it: this will only kind of help, not fully fix it. Because adding a 0/1 column just lets the model add or subtract a flat dollar amount for rich houses. It still cant change the slope for sqft. So the price per extra square foot stays the same for both groups, which is not correct.

```python
# is_rich is just 0 or 1
# price = w1*sqft + w2*bedrooms + w3*bathrooms + w_is_rich*is_rich + b
# w_is_rich only shifts things up or down by a flat amount, it does not change w1 (the sqft slope)
```

I tested it and got RMSE = $73,613. That is better than $91,347 but still way worse than $23,069. So my guess was right, adding is_rich only helps a little because it only shifts the price up, it doesn't fix the slope problem.

So then I tried adding interaction features. This means multiplying is_rich by sqft, is_rich by bedrooms, and is_rich by bathrooms. This lets the model learn an extra slope just for rich houses, on top of the normal slope.

```python
# is_rich * sqft equals sqft for rich houses, and 0 for poor houses
# this lets the model add an extra rich-only slope on top of the normal slope
features = [sqft, bedrooms, bathrooms, is_rich, is_rich*sqft, is_rich*bedrooms, is_rich*bathrooms]
```

With the interaction features added, I got RMSE = $22,956. That is basically the same as the two separate models number of $23,069.

So my conclusion is that making two separate models and making one model with the right extra features are basically the same thing. What actually matters is not how many models you use, its whether your features let the model represent what is really going on for every group.
